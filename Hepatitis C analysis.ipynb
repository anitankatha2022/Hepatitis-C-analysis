```python
dataset_uniques = data.nunique() 
dataset_uniques
```

```python
bin_vals = list(dataset_uniques[dataset_uniques == 2].index) # Create alist for our bin_vales (composed of at least two values)
bin_vals
```

```python
categorical_vals  = list(dataset_uniques[(dataset_uniques > 2) & (dataset_uniques <=6)].index)
# This function allow us to determine how many categorical do we have actually those are more than 2 and less than 6
categorical_vals
```

```python
from sklearn.preprocessing import LabelBinarizer, LabelEncoder
lb , le = LabelBinarizer(), LabelEncoder()
for col in categorical_vals: 
    data[col] = le.fit_transform(data[col])
for col in bin_vals: 
    data[col] = lb.fit_transform(data[col])
```

```python
data.head()
```

```python
X = data
y = data['Category']
```

```python
cols = X.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
X = ms.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns=[cols])
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(X.corr(),annot=True,cmap='crest')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
```

```python
# Get the inertia: 
kmeans.inertia_
```

```python
pred = kmeans.predict(X)
```

```python
data_with_clusters = data.copy()
data_with_clusters['clusters'] = pred
sns.scatterplot(x=data_with_clusters['AST'], y=data_with_clusters['ALT'],c=data_with_clusters['clusters'])
plt.xlabel('AST levels')
plt.ylabel('ALT levels')
plt.title('Making clusters')
```

```python
# value of clusters k = 10:
k_means = KMeans(n_clusters=10, random_state=20)
k_means.fit(X)
```

```python
k_means.inertia_
```

```python
pred_3 = k_means.predict(X)
```

```python
data_with_clusters = data.copy()
data_with_clusters['clusters'] = pred_3
sns.scatterplot(x=data_with_clusters['AST'], y=data_with_clusters['ALT'],c=data_with_clusters['clusters'])
plt.xlabel('AST levels')
plt.ylabel('ALT levels')
plt.title('Making a clustering when k=8')
```
